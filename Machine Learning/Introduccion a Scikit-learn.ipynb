```python
import seaborn as sns; sns.set()
%matplotlib inline
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris,hue='species',height=1.5)
```

```python
X_iris=iris.drop('species',axis=1)
```

```python
y_iris=iris['species']
```

### API Estimator y ejemplos de Machine Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng=np.random.RandomState(42)
x=10*rng.rand(50)
y=2*x+rng.randn(50)
#colors=rng.rand(50)
plt.scatter(x,y,cmap='viridis')
```

### Pasos a realizar
* Elija una clase modelo

```python
from sklearn.linear_model import LinearRegression
```

* Elija los Hiperparametros del modelo

```python
model=LinearRegression(fit_intercept=True)
model
```

* Ajustar datos de una matriz de caracteristicas y vector de destino

```python
X=x[:,np.newaxis]
X.shape
```

* Ajustar el modelo a nuestros datos

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

* Predecir etiquetas de los datos desconocidos

```python
xfit=np.linspace(-1,11)
Xfit=xfit[:,np.newaxis]
```

```python
yfit=model.predict(Xfit)
```

* ahora vamos a visualizar los resultados

```python
plt.scatter(x,y,color='red')
plt.plot(xfit,yfit,'-g')
```

## Prediccion Botanica
* Para ello vamos a seleccionar datso de entrenamiento y testeo

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest =train_test_split(X_iris,y_iris,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(Xtrain,Ytrain)
y_model=model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Ytest,y_model)
```
