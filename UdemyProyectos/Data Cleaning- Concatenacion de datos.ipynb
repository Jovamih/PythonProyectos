### Haremos un recorrido de los archivos de un directorio

```python
def ls(path=os.getcwd()):
    return [arch.path for arch in os.scandir(path) if (arch.is_file() and arch.name.endswith('csv'))]
```

```python
a=[os.path.abspath(arch) for arch in os.listdir(sub_path) if os.path.isfile(os.path.join(sub_path,arch))]
```

```python
files=ls(sub_path)
files.sort()
files
```

```python
def fullconcat(path=os.getcwd(),data=None):
    files=ls(path); files.sort()
    files.pop(0)
    for file_path in files:
        data1=pd.read_csv(file_path)
        data=pd.concat([data,data1],axis=0)
            
    return data
            
```

```python
datapro=fullconcat(sub_path,data)
datapro.shape
```
