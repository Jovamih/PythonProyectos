## Revision de datos faltantes

Se puede haceer uso de las funciones Pandas integradas
que on fillna(), dropna()

```python
import pandas as pd
import numpy as np
import random

rng=np.random.RandomState(0)
matriz=np.random.randint(10,2020,size=(10,12))
columns=list('ABCDEFGHIJKH')

data= pd.DataFrame(matriz,index=range(10),columns=columns)
data
```

```python
#Ahora haremos una indexaccion elegante para obtener valores np.nan aleatorios

data.iloc[[0,2,5,7,3],[0,3,4,4,2,4,4]]=np.nan
data
```

```python
#ahora operamos con los datos vacios o nulos -> np.nan
#borrams datos
    #que poseen una ffila con NaN
data=data.dropna(axis='rows',how='any')
data

```

Podemos rellenar los datos egun sea nuestro proposito

```python
data['D']=data['D'].fillna(data['D'].mean())
data
```

```python

for i in range(len(data.columns)):
    data.iloc[:,i]=data.iloc[:,i].fillna(data.iloc[:,i].mean())
```

### Las variables Dummy
Las variables dummy tiene una importtante aplicacion

```python
def dummies(df,column):
    dum=pd.get_dummies(df[column],prefix=column)
    df=df.drop(column,axis=1)
    return pd.concat([df,dum],axis=1)
```
