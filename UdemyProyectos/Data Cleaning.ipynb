# Carga de datos por la funcion Open

### Primero tenemos que crear un diccionario con claves igual al nombre de cada columna y valor el contenido de dichas filas
### Luego agregarlos al DataFrame

## Este procesamiento es util cuando se manejan datos de gran tama√±o

```python
data_dict=dict()
columns=list()
with open(path,'r') as f:
    columns=f.readline().strip().split(',')
    ncols=len(columns)
    data_dict={ col :list() 
               for col in columns}
    for line in f:
        datos=line.strip().split(',')
        for i in range(ncols):
            valor=datos[i] if len(datos[i])!=0 else np.nan
            data_dict[columns[i]].append(valor)
# luego de haber obtenido las columnas en un diccionario, estamos listos para rellenarlos

print(data_dict)
```

```python
data=pd.DataFrame(data_dict)
data
```
